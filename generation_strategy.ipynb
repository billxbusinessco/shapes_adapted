```python
import pygad
import numpy as np

def fitness_func(ga_instance,solution, solution_idx):
    normalized_weights = np.array(solution)
    normalized_weights = normalized_weights/normalized_weights.sum()

    lol = first_analysis.weighted_generation_parallel(weight = normalized_weights, seed_points = [3,8])
    spectra = np.concatenate(lol)
    minimas = first_analysis.find_min_dataset(spectra)
    mean_minimas = np.mean(minimas)
    var_minimas = np.var(minimas)
    return mean_minimas

def on_start(ga_instance):
    for i in range(ga_instance.sol_per_pop):
        ga_instance.population[i] = np.random.dirichlet(np.ones(ga_instance.num_genes))

def on_generation(ga_instance):
    global last_fitness
    best_solution, best_fitness, _ = ga_instance.best_solution(pop_fitness=ga_instance.last_generation_fitness)
    
    print(f"Generation = {ga_instance.generations_completed}")
    print(f"Fitness    = {best_fitness}")
    print(f"Change     = {best_fitness - last_fitness}")
    print(f"Genes      = {best_solution}")
    
    last_fitness = best_fitness

mutation_percent_genes=30  # or adjust based on num_genes
num_genes=7
ga_instance = pygad.GA(
    num_generations=5,
    num_parents_mating=5,
    fitness_func=fitness_func,
    sol_per_pop=10,
    num_genes=num_genes,
    init_range_low=0.2,
    init_range_high=1,
    on_start=on_start,
    gene_space={'low': 0.2, 'high': 1},
    mutation_percent_genes=mutation_percent_genes,
    on_generation=on_generation,
    keep_elitism=2
)

ga_instance.run()

# Best solution
solution, fitness, _ = ga_instance.best_solution()
print("Best:", solution)
print("Sum:", np.sum(solution))
```

## Genetic selection with database

##### let'ts take inspiration from the best weights we got from our genettic category weight selection to maximize difference

```python
weights = np.array([0.02065873,0.12101462,0.08081627,0.11466537,0.30777906,0.44471586])

normalized_weights = np.array(weights)
normalized_weights = normalized_weights/normalized_weights.sum()

lol = first_analysis.weighted_generation_parallel(number_of_samples=40000, weight = normalized_weights, seed_points = [3,9])
```

```python
final_spectra  = np.concatenate(lol)
```

```python
final_spectra = final_spectra[np.random.permutation(final_spectra.shape[0]), :]
```

```python
import pygad
import numpy as np

rarg_spectra = final_spectra.reshape(1000,40,-1)
def fitness_func(ga_instance,solution, solution_idx):
    global rarg_spectra
    axis_0 = np.arange(0,1000)
    axis_1 = solution

    selection = rarg_spectra[axis_0,axis_1,:]
    minimas = first_analysis.find_min_dataset(selection)
    mean_minimas = np.mean(minimas)
    var_minimas = np.var(minimas)
    return mean_minimas



def on_generation(ga_instance):
    best_solution, best_fitness, _ = ga_instance.best_solution(pop_fitness=ga_instance.last_generation_fitness)
    
    print(f"Generation = {ga_instance.generations_completed}")
    print(f"Fitness    = {best_fitness}")
    print(f"Genes      = {best_solution}")
    

mutation_percent_genes = 50  # starting value
num_genes = 1000
gene_space_array = np.arange(0, 40, dtype=int)

ga_instance = pygad.GA(
    num_generations=100,
    num_parents_mating=20,
    fitness_func=fitness_func,
    sol_per_pop=100,
    num_genes=num_genes,
    gene_space=gene_space_array,
    mutation_percent_genes=mutation_percent_genes,  # initial value
    on_generation=on_generation,
    keep_elitism=2,
    gene_type=int,
    parent_selection_type="tournament",
    K_tournament=3
)


ga_instance.run()

# Best solution
solution, fitness, _ = ga_instance.best_solution()
print("Best:", solution)
print("Sum:", np.sum(solution))
```
